## Merge column-wise into single dataframe

```python
gsr_df = pd.concat([bond_ratings_df,fund_allocations_df,fund_ratios_df,fund_specs_df,other_specs_df,
                    return_3year_df,return_5year_df,return_10year_df],axis=1)
```

```python
print(gsr_df.shape)
gsr_df.tail()
#.transpose()
```

```python
# remove duplicate columns
gsr_df = gsr_df.loc[:,~gsr_df.columns.duplicated()]

gsr_df.astype(float)
```

```python
#Apply z-score
gsr_df = gsr_df.apply(zscore)

gsr_df.tail()
```

```python
#Check for missing values
print(gsr_df.isna().sum())

#Getting 5-point summary
gsr_df.describe()
```

```python
# Append previously saved columns
gsr_df['fund_id'] = fund_id
gsr_df['greatstone_rating'] = greatstone_rating
```

```python
#Check for missing values
print(gsr_df.isna().sum())

#Getting 5-point summary
gsr_df.describe().transpose()
```

## Divide Train and Test Data

## Train Set -> Rows where greatstone_rating is not null

```python
train_df = gsr_df[gsr_df['greatstone_rating'].notna()]
print(train_df.shape)
train_df.head()
#.transpose()
```

```python
X_train = train_df.drop(['fund_id','greatstone_rating'], axis=1)
y_train = train_df['greatstone_rating'].astype(int)
```

```python
X_train.describe().transpose()
```

## Test Set -> Rows where greatstone_rating is null

```python
test_df = gsr_df[gsr_df['greatstone_rating'].isna()]
print(test_df.shape)
test_df.head()
#.transpose()
```

```python
X_test = test_df.drop(['fund_id','greatstone_rating'], axis=1)
y_test = test_df['greatstone_rating']
```

```python
X_test.describe()
```

## SVC - Train and Predict

```python
#Train and Predict using SVM
from sklearn.svm import SVC

svc = SVC(kernel='rbf', C=10, gamma=1000)
#SVC(gamma='auto')
svc.fit(X_train, y_train)
```

```python
# predict the response
y_pred = svc.predict(X_test)
```
